```
import pandas as pd  # Import pandas

# Creating the table as a DataFrame
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"],
    "Cookies Sold": [50, 55, 52, 60, 65, 70]
}

df = pd.DataFrame(data)  # Convert dictionary into a DataFrame

# Display the table
df
```

```
high_cookie_sales = df[df["Cookies Sold"] > 60]  # Filter where Cookies Sold > 60

# Display the days
print("Days where more than 60 cookies were sold:")
print(high_cookie_sales["Day"].tolist())  # Convert to list for easy reading

```

```
import pandas as pd  # Importing the pandas library

# Creating the table as a DataFrame
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"],
    "Pies Sold": [5, 6, 7, 10, 15, 18]
}

df = pd.DataFrame(data)  # Converting dictionary into a DataFrame

# Display the table
df
```

```
max_pies_day = df.loc[df["Pies Sold"].idxmax(), "Day"]  # Get the day with max pies sold

print(f"The highest number of pies was sold on: {max_pies_day}")
```
